```python
import pandas as pd
from collections import defaultdict
from tqdm import tqdm
import pickle
```

```python
taobao = pd.read_csv('./data/ijcai2016_taobao.csv')
```

```python
freq = defaultdict(dict)
for i,row in tqdm(taobao.iterrows()):
    use_ID = row['use_ID']
    sel_ID = row['sel_ID']
    if sel_ID in freq[use_ID]:
        freq[use_ID][sel_ID] += 1
    else:
        freq[use_ID][sel_ID] = 1
```

```python
freq_list = {}
for key, lst in tqdm(freq.items()):
    freq_list[key] = sorted(list(lst.items()), key = lambda x: x[1], reverse = True)
```
