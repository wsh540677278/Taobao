```python
import pandas as pd
from tqdm import tqdm
from sklearn.cluster import SpectralClustering
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing 
from collections import defaultdict
import warnings
warnings.filterwarnings("ignore")
from sklearn.datasets import make_biclusters
import matplotlib.pyplot as plt
import seaborn as sns
from copy import deepcopy
import collections
import pickle
sns.set(color_codes=True)
```

```python
df1 = pd.DataFrame(pd.read_csv('./data/ijcai2016_koubei_train.csv',header=0))
```

```python
df1[:5]
```

```python
test_data_map = {}
for i in tqdm(range(len(df1))):
    if df1.iloc[i].use_ID in test_data_map:
        if df1.iloc[i].loc_ID in test_data_map[df1.iloc[i].use_ID]:
            test_data_map[df1.iloc[i].use_ID][df1.iloc[i].loc_ID].append(df1.iloc[i].mer_ID)
        else:
            test_data_map[df1.iloc[i].use_ID][df1.iloc[i].loc_ID]=[df1.iloc[i].mer_ID]
    else:
        test_data_map[df1.iloc[i].use_ID] = {}
        test_data_map[df1.iloc[i].use_ID][df1.iloc[i].loc_ID]=[df1.iloc[i].mer_ID]
```

```python
test_data_map
```

```python
sorted_test_data_map = {}
for i in tqdm(test_data_map):
    sorted_test_data_map[i] = {}
    for j in test_data_map[i]:
        sorted_test_data_map[i][j] = {}
        
for i in tqdm(test_data_map):
    for j in test_data_map[i]:
        sorted_test_data_map[i][j] = sorted(collections.Counter(test_data_map[i][j]).items(), key=lambda x: x[1], reverse=True)
         
        
```
