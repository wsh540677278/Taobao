```python
import pandas as pd
from collections import defaultdict
from tqdm import tqdm
import pickle
```

```python
koubei = pd.read_csv('./data/ijcai2016_koubei_train')
```

```python
new_old_id_maps = pickle.load(open('./data/new_old_id_maps.pkl',"rb"))
```

```python
mer_new_old_id_maps = new_old_id_maps['mer_new_old_id_maps']
mer_old_new_id_maps = new_old_id_maps['mer_old_new_id_maps']
```

```python
freq = defaultdict(dict)
for i,row in tqdm(koubei.iterrows()):
    use_ID = row['use_ID']
    mer_ID = row['mer_ID']
    loc_ID = row['loc_ID']
    if (mer_ID, loc_ID) in freq[use_ID]:
        freq[use_ID][(mer_ID, loc_ID)] += 1
    else:
        freq[use_ID][(mer_ID, loc_ID)] = 1
```

```python
freq_list = {}
for key, lst in tqdm(freq.items()):
    freq_list[key] = sorted(list(lst.items()), key = lambda x: x[1], reverse = True)
```

```python
freq_list
```

```python
new_freq_list = {}
for key, lst in tqdm(freq_list.items()):
    if key not in user_old_new_id_maps:
        continue
    new_use_id = user_old_new_id_maps[key]
    new_freq_list[new_use_id] = []
    for (old_mer_id, loc_id), freq in lst:
        new_mer_id = mer_old_new_id_maps[(old_mer_id, loc_id)]
        new_freq_list[new_use_id].append((new_mer_id, freq))
```

```python
len(new_freq_list)
```

```python
pickle.dump(new_freq_list, open('./data/user_merchant_freq_new_id.pkl', 'wb'))
```
