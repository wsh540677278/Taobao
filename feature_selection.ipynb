```python
merchant_count = last_week['mer_ID'].value_counts()
```

```python
seller_count = a_week['sel_ID'].value_counts()
```

```python
seller_merchant = a_week.merge(last_week, left_on = 'use_ID', right_on = 'use_ID', how = 'inner')
```

```python
seller_merchant_count = defaultdict(int)
for index, row in tqdm(seller_merchant.iterrows()):
    seller_merchant_count[(row['sel_ID'], row['mer_ID'])] += 1
```

```python
seller_merchant_sum = 0
for v in seller_merchant_count.values():
    seller_merchant_sum += v
for k, v in seller_merchant_count.items():
    seller_merchant_count[k] = v/seller_merchant_sum
```

```python
def prob(dic):
    s = 0
    for v in dic.values():
        s += v
    for k, v in dic.items():
        dic[k] = v/s
    return dic
```

```python
merchant_count_dic = prob(defaultdict(float, dict(merchant_count.items())))
seller_count_dic = prob(defaultdict(float, dict(seller_count.items())))
```

```python
print(len(merchant_count_dic))
print(len(seller_count_dic))
print(len(seller_merchant_count))
```

```python
lift = defaultdict(float)
for merchant in tqdm(merchant_count_dic):
    for seller in seller_count_dic:
        if (seller, merchant) in seller_merchant_count:
            lift[(seller, merchant)] = seller_merchant_count[(seller, merchant)] / (merchant_count_dic[merchant] * seller_count_dic[seller])
```

```python
lift = sorted(lift.items(), key = lambda x: x[1], reverse = True)
```
